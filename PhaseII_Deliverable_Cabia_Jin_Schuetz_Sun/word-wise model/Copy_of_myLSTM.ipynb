```
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
h = np.load('data/hist.npy')
labels = []
values = []
for i,j in h.tolist().items():
  labels.append(i)
  values.append(j)
plt.plot(range(50),values[0],range(50),values[1],range(50),values[2],range(50),values[3])
plt.xlabel('epochs')
plt.legend(labels=['val_loss', 'val_accuracy', 'loss', 'accuracy'])
plt.show()
```
